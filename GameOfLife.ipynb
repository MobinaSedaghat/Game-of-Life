```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from matplotlib.colors import LinearSegmentedColormap
import matplotlib as mpl
import random
```

```python
class Board():

  def __init__(self, rows, cols, num):
    self.rows = rows
    self.cols = cols
    self.num = num
    self.grid = np.zeros((self.rows,self.cols))

    rand1 =[random.randint(0,self.rows-1) for i in range(random.randint(0, self.cols * self.rows))]
    rand2 = [random.randint(0,self.cols-1) for i in range(len(rand1))]


    for i in range (len(rand1)):
      self.grid[rand1[i],rand2[i]] = 1



  def update(self,*argue):
    self.im.set_data(self.grid)

    if self.gen_num > 0 :
      new_grid = GameOfLife.next_gen(self)
      self.grid = new_grid

    suf = lambda n: "%d%s"%(n,{1:"st",2:"nd",3:"rd"}.get(n if n<20 else n%10,"th"))
    plt.title(str(suf(self.gen_num)) + " generation" , fontsize = 30)
    self.gen_num = self.gen_num + 1

    return [self.im]


  def draw_grid(self):
    self.gen_num = 0

    fig,ax = plt.subplots(figsize=(10,10))

    #grid lines
    ax = plt.gca()
    ax.set_xticks(np.arange(-.5, self.grid.shape[1], 1))
    ax.set_yticks(np.arange(-.5, self.grid.shape[0], 1))
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    plt.ylabel('Number of rows: '+ str(self.rows))
    plt.xlabel('Number of columns: '+ str(self.cols))
    #ax.grid(color='black', linestyle='-', linewidth=2)

    cmap = mpl.colors.ListedColormap(['#727272', '#CF9FFF'])
    self.im = ax.imshow(self.grid,cmap= cmap)

    self.ani = FuncAnimation(fig, self.update ,frames= self.num ,interval = 300)

    html = HTML(self.ani.to_html5_video())
    display(html)
    plt.close()
```

```python
class GameOfLife(Board):

  def count_neighbors(self):

    self.neighs = np.zeros((len(self.grid[:,0]) , len(self.grid[0,:])))

    self.grid=np.append(self.grid,np.zeros(( len(self.grid[:,0]), 1)), axis =1)
    self.grid=np.append(self.grid, np.zeros(( 1, len(self.grid[0,:]))), axis=0)

    self.neighs[0,0] = self.grid[0,1] + self.grid [1 , 0] + self.grid [1,1]

    for i in range(1,self.rows):
      self.neighs[i,0] = np.sum(self.grid[i-1:i+2 , 0:2]) - self.grid[i,0]

      for j in range(1,self.cols):
        self.neighs[0,j] = np.sum(self.grid[0:2 , j-1:j+1]) - self.grid[0,j]
        self.neighs[i,j] = np.sum(self.grid[i-1:i+2 , j-1:j+2]) - self.grid[i,j]
    return self.neighs

  def next_gen(self):
    GameOfLife.count_neighbors(self)
    nonzero_ind = np.nonzero(self.neighs)

    next_grid = np.zeros(self.rows * self.cols ).reshape(self.rows,self.cols)

    for i,j in zip(nonzero_ind[0],nonzero_ind[1]):
      if self.grid[i,j] and (self.neighs[i,j] < 2 or self.neighs[i,j] > 3):
        next_grid[i,j] = 0

      if self.grid[i,j]  and (self.neighs[i,j] == 2 or self.neighs[i,j] == 3):
        next_grid[i,j] = 1

      if not(self.grid[i,j]) and self.neighs[i,j] == 3:
        next_grid[i,j] = 1

    return next_grid
```
